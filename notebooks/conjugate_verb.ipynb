```python
from pygramadan.printer_neid import PrinterNeid
from pygramadan.verb import Verb
from pathlib import Path
```

```python
#@title Enter verb

verb = 'aimsigh' #@param {type:"string"}
```

```python
verb_dir = Path('/content/BuNaMo/verb')
verb_pat = verb + '_' + '*.xml'
for verb_file in verb_dir.glob(verb_pat):
    v = Verb(source=verb_file)
    pn = PrinterNeid(with_xml_declarations=True)
    out = pn.print_verb_xml(v)
    f = open(f"{verb}_print.xml", "wb")
    f.write(out)
```

```python
%%capture
!wget 'https://raw.githubusercontent.com/michmech/Gramadan/master/NeidOutput/!gram.xsl'
```
